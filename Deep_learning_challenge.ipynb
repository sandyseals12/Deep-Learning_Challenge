```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
application_df = application_df.drop(columns=['EIN', 'NAME'])
application_df
```

```python
unique_counts = application_df.nunique()
unique_counts
```

```python
application_type_counts = application_df['APPLICATION_TYPE'].value_counts()
application_type_counts
```

```python
cutoff_value = 1000
application_types_to_replace = []
application_type_counts = application_df['APPLICATION_TYPE'].value_counts()
for app_type, count in application_type_counts.items():
    if count < cutoff_value:
        application_types_to_replace.append(app_type)
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app, 'Other')
application_df['APPLICATION_TYPE'].value_counts()
```

```python
class_counts = application_df['CLASSIFICATION'].value_counts()
print(class_counts)
```

```python
class_counts = application_df['CLASSIFICATION'].value_counts()
class_counts_greater_than_1 = class_counts[class_counts > 1]
print(class_counts_greater_than_1)
```

```python
cutoff_value = 1000
classifications_to_replace = []
classification_counts = application_df['CLASSIFICATION'].value_counts()
for cls, count in classification_counts.items():
    if count < cutoff_value:
        classifications_to_replace.append(cls)
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls, 'Other')
classification_value_counts = application_df['CLASSIFICATION'].value_counts()
print(classification_value_counts)
```

```python
categorical_columns = ['APPLICATION_TYPE', 'AFFILIATION', 'USE_CASE', 'ORGANIZATION', 'INCOME_AMT', 'SPECIAL_CONSIDERATIONS']
application_df = pd.get_dummies(application_df, columns=categorical_columns)
```

```python
X= application_df.drop(columns=['IS_SUCCESSFUL'])
y= application_df['IS_SUCCESSFUL']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
print(X_train.columns)
```

```python
non_numeric_columns = ['EIN', 'NAME', 'CLASSIFICATION', 'STATUS']
X_train_numeric = X_train.drop(columns=non_numeric_columns)
X_test_numeric = X_test.drop(columns=non_numeric_columns)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_numeric)
X_test_scaled = scaler.transform(X_test_numeric)
nn = tf.keras.models.Sequential()
input_units = X_train_scaled.shape[1]
nn.add(tf.keras.layers.Dense(units=64, input_dim=input_units, activation='relu'))
nn.add(tf.keras.layers.Dense(units=32, activation='relu'))
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
epochs = 50
nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
nn.fit(X_train_scaled, y_train, epochs=epochs, verbose=1)
nn.summary()
```

```python
model_loss, model_accuracy = nn.evaluate(X_test_scaled, y_test, verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
